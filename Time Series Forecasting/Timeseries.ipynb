### **What is Time Series Forecasting?**

**Time Series Forecasting** is a method used to predict future values based on previously observed data points that are ordered in time. Unlike other forms of prediction that might deal with unordered data, time series forecasting explicitly considers the order of data points and often the time intervals between them. 

Time series data is unique in that the value of a data point depends not only on the current inputs (like in typical regression) but also on its previous values and the order in which they occur. This makes time series forecasting particularly suitable for tasks where the temporal aspect is crucial, such as weather prediction, stock market analysis, sales forecasting, and demand prediction.

### **Example of Time Series Forecasting:**

Imagine you have monthly sales data for a retail store for the past five years. You want to predict the sales for the next 12 months. Time series forecasting techniques like ARIMA (AutoRegressive Integrated Moving Average) can be used to model the historical sales data and predict future sales, considering trends, seasonality, and other temporal factors.

### **How Time Series Forecasting Differs from Regression:**

**1. **Nature of Data:**
   - **Time Series Forecasting**: Deals specifically with data that is collected over time and where the order of observations is critical. The future value depends on the historical data points.
   - **Regression**: Generally focuses on predicting a target variable based on one or more predictor variables, without necessarily considering the order of data points.

**2.** **Autocorrelation:**
   - **Time Series Forecasting**: Involves autocorrelation, where past values influence future values. This is a key aspect of time series models.
   - **Regression**: Typically assumes that the observations are independent of each other, meaning that past values do not influence future predictions.

**3.** **Purpose:**
   - **Time Series Forecasting**: Used for predicting future values that are dependent on past data, often capturing trends, seasonality, and cycles.
   - **Regression**: Used for understanding relationships between variables and predicting the value of a dependent variable from one or more independent variables.

**4.** **Modeling Approach:**
   - **Time Series Forecasting**: Utilizes models like ARIMA, SARIMA, Exponential Smoothing, etc., that are specifically designed to handle time-dependent data.
   - **Regression**: Utilizes models like Linear Regression, Polynomial Regression, etc., where the focus is on fitting a curve or line to the data.

   In summary, **time series forecasting** is specialized for dealing with sequential data where past values influence future predictions, whereas **regression** is more general and is used to model relationships between variables without necessarily considering their order.


### **Exploratory Data Analysis (EDA) of Time Series Data**

Exploratory Data Analysis (EDA) for time series data involves understanding the underlying patterns in the data. The main aspects to focus on during EDA for time series are **trends**, **seasonality**, and **anomalies**. Let’s dive into each of these concepts in detail.

#### **1. Trend**
- **Definition:** A trend in time series data is the long-term movement or direction in the data over a period of time. It indicates the general tendency of the data to increase, decrease, or remain stable over time.
- **Types of Trends:**
  - **Upward Trend:** When the data shows a general increase over time.
  - **Downward Trend:** When the data shows a general decrease over time.
  - **No Trend:** When the data does not show any clear upward or downward movement.
- **Why It Matters:** Identifying trends helps in understanding the long-term direction of the data, which is crucial for making future predictions. For example, if a company’s sales data shows an upward trend, it may indicate growth and help in planning for future resource allocation.
- **Visualizing Trends:**
  - **Line Plot:** A simple line plot of the data over time can reveal the trend.
  - **Moving Average:** Applying a moving average smooths out short-term fluctuations and helps highlight the trend.


#### **2. Seasonality**
- **Definition:** Seasonality refers to patterns that repeat at regular intervals over time. These patterns are often linked to seasonal factors like weather, holidays, or economic cycles.
- **Examples of Seasonality:**
  - **Monthly Seasonality:** Retail sales often increase during the holiday season every year.
  - **Daily Seasonality:** Website traffic might peak during certain hours of the day.
- **Why It Matters:** Recognizing seasonality is important for making accurate predictions, as it allows you to account for periodic fluctuations. For instance, if a retailer knows that sales spike every December, they can stock up on inventory accordingly.
- **Visualizing Seasonality:**
  - **Seasonal Decomposition:** Decomposing the time series into its trend, seasonal, and residual components helps isolate the seasonal effect.
  - **Box Plots:** Box plots grouped by month or season can reveal seasonality in the data.


  #### **3. Anomalies**
- **Definition:** Anomalies (or outliers) in time series data are data points that deviate significantly from the expected pattern. These could be due to rare events, errors in data collection, or sudden changes in the system being observed.
- **Types of Anomalies:**
  - **Point Anomalies:** A single data point is far outside the range of expected values.
  - **Contextual Anomalies:** A data point that is anomalous within a specific context (e.g., a high sales spike in an off-season).
  - **Collective Anomalies:** A sequence of data points that collectively deviate from the norm.
- **Why It Matters:** Detecting anomalies is critical in many applications such as fraud detection, fault detection in machinery, and monitoring network traffic. Anomalies can indicate significant but often unexpected events that require further investigation.
- **Visualizing Anomalies:**
  - **Line Plot with Markers:** Highlighting the anomalies on a line plot can make them easier to spot.
  - **Z-Score Analysis:** Using statistical methods like the Z-score to identify outliers.


### **Summary of EDA in Time Series**

1. **Trend:** Helps identify the overall direction of the data over time, whether it's increasing, decreasing, or remaining constant. It can be visualized through line plots and moving averages.

2. **Seasonality:** Captures regular and periodic fluctuations in data, often influenced by seasonal factors. It can be visualized through seasonal decomposition or box plots.

3. **Anomalies:** Highlights unexpected or unusual data points that deviate from the norm. Anomalies can be detected and visualized using methods like Z-score analysis.

### **Moving Average

Moving Average (MA) is a widely used statistical technique in time series analysis and forecasting. It is a method to smooth out short-term fluctuations in the data and highlight longer-term trends or cycles. Moving averages are particularly useful in reducing noise and making it easier to spot underlying patterns.

**Types of Moving Averages**
There are several types of moving averages, each with its own method of calculation and application:

1. **Simple Moving Average (SMA)**: The Simple Moving Average is the unweighted mean of the previous 
n data points. It is called "moving" because it is calculated for successive subsets of data points.


2. **Weighted Moving Average (WMA)** : The Weighted Moving Average gives more importance to recent data points, unlike SMA where all points have equal weight. The weights decrease linearly as the observations move further back in time.

3. **Exponential Moving Average (EMA)** : The Exponential Moving Average assigns exponentially decreasing weights to older data points, making it more responsive to recent changes in the data compared to SMA and WMA.

```python

import pandas as pd
import matplotlib.pyplot as plt


# url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv("airline-passengers.csv", parse_dates=['Month'], index_col='Month')

# Perform EDA
print(data.head())
print(data.describe())

# Plot the data
plt.figure(figsize=(12,6))
plt.plot(data, label='AirPassengers')
plt.title('AirPassengers Data')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
# Calculate the moving average
data['Moving_Average'] = data['Passengers'].rolling(window=12).mean()

# Plot the moving average
plt.figure(figsize=(12,6))
plt.plot(data['Passengers'], label='Original')
plt.plot(data['Moving_Average'], color='red', label='12-Month Moving Average')
plt.title('Moving Average of AirPassengers Data')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

### **ARIMA: Autoregressive Integrated Moving Average**

ARIMA, short for **Autoregressive Integrated Moving Average**, is a powerful and popular statistical modeling technique used for time series forecasting. It is particularly useful for understanding and predicting future points in a time series by considering the data's own past values, its trend, and seasonal patterns. 

ARIMA is composed of three main components:

1. **Autoregressive (AR)** : The AR component involves regressing the time series on its own past values. The number of lagged observations included in the model is defined by the parameter. The AR component captures the relationship between the current value and its previous values. For instance, if p = 2 , the model uses the last two observations to predict the current value.

2. **Integrated (I)**: The I (Integrated) component deals with the stationarity of the time series. A time series is said to be stationary if its statistical properties like mean, variance, and autocorrelation are constant over time. The integrated component involves differencing the time series to make it stationary. The number of times the differencing is applied is defined by the parameter. Differencing is used to remove trends and make the series stationary. For example, if d = 1, the first difference of the series is used.

3. **Moving Average (MA)**: The MA component involves modeling the error of the time series as a linear combination of past error terms. The number of lagged forecast errors included in the model is defined by the parameter.The MA component captures the relationship between the current value and the past errors. For instance, if q = 2, the model uses the last two errors to predict the current value.

The ARIMA model is formed by combining the AR, I, and MA components. The model is specified by three parameters: \( p \), \( d \), and \( q \). These parameters represent the order of the AR, I, and MA components, respectively.

- **ARIMA(p, d, q):**
  - \( p \) = Number of lag observations included in the model (AR).
  - \( d \) = Number of times that the raw observations are differenced (I).
  - \( q \) = Size of the moving average window (MA).

### **Steps to Build an ARIMA Model**

1. **Visualize the Time Series:**
   - Plot the time series data to identify patterns, trends, seasonality, and stationarity.

2. **Make the Series Stationary:**
   - Use differencing to make the time series stationary. Check stationarity using statistical tests like the Augmented Dickey-Fuller (ADF) test.

3. **Identify the Order of the ARIMA Model:**
   - Use Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots to identify the appropriate values of p , d, and q.

4. **Fit the ARIMA Model:**
   - Fit the ARIMA model using the identified parameters.

5. **Make Predictions:**
   - Use the fitted model to make forecasts and evaluate its performance.



```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Load data
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Step 1: Visualize the Time Series
plt.figure(figsize=(12,6))
plt.plot(data['Passengers'], label='Original Data')
plt.title('AirPassengers Data')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# !pip install statsmodels
```

```python
#Make the Series Stationary (Differencing)
data['Differenced'] = data['Passengers'].diff().dropna()

data.head()
```

```python
# Check stationarity using ADF test
result = adfuller(data['Differenced'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
#Identify the Order of the ARIMA Model (p, d, q)
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plt.figure(figsize=(12,6))

plt.subplot(121)
plot_acf(data['Differenced'].dropna(), lags=20, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')

plt.subplot(122)
plot_pacf(data['Differenced'].dropna(), lags=20, ax=plt.gca(), method='ywm')
plt.title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```

```python
#Fit the ARIMA Model
model = ARIMA(data['Passengers'], order=(1,1,1))
model_fit = model.fit()

#Make Predictions
predictions = model_fit.predict(start=len(data), end=len(data)+11, typ='levels')

# Plot the original data and predictions
plt.figure(figsize=(12,6))
plt.plot(data['Passengers'], label='Original Data')
plt.plot(predictions, color='red', label='ARIMA Predictions')
plt.title('ARIMA Model Predictions')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```
